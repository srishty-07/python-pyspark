```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.array([1,10])
y_axis = np.array([0,-20])
```

```python
plt.plot(x_axis,y_axis)
```

```python
y_points=np.array([2,-7,10,13,26])
plt.plot(y_points,marker="*",color="pink")
plt.title("sample chart")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
```

```python
# Create a list of months: ['January', 'February', 'March', 'April', 'May'].
# Create a list of monthly sales data for a hypothetical product: [2500, 3200, 2800, 4100, 3700]
months=['January', 'February', 'March', 'April', 'May']
sales=[2500, 3200, 2800, 4100, 3700]
#create a bar chart 
#lable the axes month(x) and sales(y_axis) lable the title :sales data

month_array = np.array(months)
sales_array = np.array(sales)
```

```python
plt.bar(month_array,sales_array,color='pink')
plt.xlabel('month')
plt.ylabel('sales')
plt.title('Monthly Sales Data')
```
